**瓶颈生成树（Bottleneck Spanning Tree）**

带有权重的无向图 $G(V,E,w)$ 的瓶颈生成树，表现为：在所有生成树中，最大权重边的权重值最小。即，BST $T$ 最小化瓶颈损失 $c(T)=max_{e \in T}{w(e)}$。

3. 设计一个线性时间复杂度的算法：对于给定的图 $G(V,E,w)$，找到其瓶颈生成树，分析算法设计思路，并基于python编程实现。

```python
#问题1：证明G的每一个最小生成树（MST）都是瓶颈生成树（BST）
# add your code here
import networkx as nx

def is_BST(G, b):
    # 将边按权重排序
    edges = sorted(G.edges(data=True), key=lambda x: x[2]['weight'])

    # 初始化生成树
    MST = nx.Graph()
    MST.add_nodes_from(G.nodes())

    # 添加边到生成树，直到最大权重边超过b或生成树包含所有顶点
    for edge in edges:
        m, n = edge[0], edge[1]
        if edge[2]['weight'] > b:
            return False
        else:
            print("")
        if not MST.has_edge(m, n):
            MST.add_edge(m, n)
            if nx.is_connected(MST):
                return True
            else:
                print("")

    return False

X = nx.Graph()
X.add_edge('A', 'B', weight=1)
X.add_edge('B', 'C', weight=2)
X.add_edge('C', 'D', weight=3)
X.add_edge('A', 'C', weight=4)
X.add_edge('B', 'D', weight=5)

Y = 3
print(is_BST(X, Y))  # 输出: True

# algorithm of the liear time complexity ：算法时间复杂度为 O(ElogE)，其中 E 是图中边的数量。
```

```python
#问题2：设计一个线性时间复杂度的算法：对于一个图 G(V,E,w)和一个整数b，判断图 G是否存在一个瓶颈生成树，其最大权重边的权重不超过b，分析算法设计思路，并基于python编程实现。
# add your idea here
#     对于这个问题，我们可以对 Kruskal 算法进行修改。Kruskal 算法是一种用于求最小生成树的算法，它按照边的权重从小到大排序，依次选择边加入生
#     成树，只要加入这条边不会形成环。首先，将所有边按照权重从小到大排序。然后，从权重小于等于 b 的边开始，依次尝试加入生成树。如果在这个过
#     程中能够成功构建一个生成树（即包含所有顶点且没有环），那么就说明存在一个瓶颈生成树的最大权重边权重不超过 b。
    
# add your code here
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.rank[root_x] > self.rank[root_y]:
                self.parent[root_y] = root_x
            elif self.rank[root_x] < self.rank[root_y]:
                self.parent[root_x] = root_y
            else:
                self.parent[root_y] = root_x
                self.rank[root_x] += 1
            return True
        return False


def has_bottleneck_spanning_tree(edges, n, b):
    # edges 是一个列表，每个元素是一个元组 (u, v, w)，表示边 (u, v) 的权重为 w
    # n 是顶点数量
    # b 是给定的权重上限
    edges.sort(key=lambda x: x[2])  # 按权重从小到大排序
    uf = UnionFind(n)
    for u, v, w in edges:
        if w <= b:
            if uf.union(u, v):
                if uf.find(0) == uf.find(n - 1):  # 假设所有顶点都在一个连通分量中
                    return True
    return False

# 示例图的边，格式为 (u, v, w)
edges = [
    (0, 1, 3),
    (0, 2, 5),
    (1, 2, 2),
    (1, 3, 6),
    (2, 3, 1),
    (3, 4, 4)
]
n = 5  # 顶点数量

# 判断是否存在瓶颈生成树的最大权重边权重不超过 b
b = 4
print(has_bottleneck_spanning_tree(edges, n, b))  # 输出 True 或 False

# algorithm of the liear time complexity 
#     对边进行排序的时间复杂度是 O(ElogE)，其中E是边的数量。但是题目要求线性时间复杂度，所以不能直接排序。由于边的权重是整数，我们可以使用计
#     数排序或者桶排序等线性时间排序算法（假设边的权重范围是合理的）。这样排序的时间复杂度可以达到 O(E)。并查集操作（包括查找和合并）的摊还
#     时间复杂度是 O(α(n))，其中 α 是阿克曼函数的反函数，对于所有实际的输入规模，α(n) 可以认为是一个非常小的常数。所以并查集操作可以认为
#     是接近 O(1) 的。因此，整个算法的时间复杂度可以达到 O(E)。
```

```python
#设计一个线性时间复杂度的算法：对于给定的图 G(V,E,w)，找到其瓶颈生成树，分析算法设计思路，并基于python编程实现。
# add your idea here
#     瓶颈生成树的最大权重边的权重值是一个连续的范围，我们可以利用二分查找来确定这个最大权重值。首先，确定边的权重范围，最小权重为所有边的最
#     小权重，最大权重为所有边的最大权重。然后在这个范围内进行二分查找，每次取中间值作为 b，调用上面判断是否存在瓶颈生成树的最大权重边权重不
#     超过b的算法。如果存在瓶颈生成树的最大权重边权重不超过 b，那么说明瓶颈生成树的最大权重边权重可能更小，继续在左半部分查找；否则在右半部
#     分查找。直到找到最小的 b，使得存在瓶颈生成树的最大权重边权重不超过 b。
# add your code here
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.rank[root_x] > self.rank[root_y]:
                self.parent[root_y] = root_x
            elif self.rank[root_x] < self.rank[root_y]:
                self.parent[root_x] = root_y
            else:
                self.parent[root_y] = root_x
                self.rank[root_x] += 1
            return True
        return False

def has_bottleneck_spanning_tree(edges, n, b):
    # edges 是一个列表，每个元素是一个元组 (u, v, w)，表示边 (u, v) 的权重为 w
    # n 是顶点数量
    # b 是给定的权重上限
    edges.sort(key=lambda x: x[2])  # 按权重从小到大排序
    uf = UnionFind(n)
    for u, v, w in edges:
        if w <= b:
            if uf.union(u, v):
                if uf.find(0) == uf.find(n - 1):  # 假设所有顶点都在一个连通分量中
                    return True
    return False

def find_bottleneck_spanning_tree(edges, n):
    min_w = min(edge[2] for edge in edges)
    max_w = max(edge[2] for edge in edges)
    while min_w < max_w:
        mid = (min_w + max_w) // 2
        if has_bottleneck_spanning_tree(edges, n, mid):
            max_w = mid
        else:
            min_w = mid + 1
    # 最终 min_w 就是瓶颈生成树的最大权重边的权重
    return min_w

# 示例图的边，格式为 (u, v, w)
edges = [
    (0, 1, 3),
    (0, 2, 5),
    (1, 2, 2),
    (1, 3, 6),
    (2, 3, 1),
    (3, 4, 4)
]
n = 5  # 顶点数量

# 判断是否存在瓶颈生成树的最大权重边权重不超过 b
b = 4
print(has_bottleneck_spanning_tree(edges, n, b))  # 输出 True 或 False

# 找到瓶颈生成树
print(find_bottleneck_spanning_tree(edges, n))  # 输出瓶颈生成树的最大权重边的权重

# algorithm of the liear time complexity 
#     二分查找的次数最多为 O(log(max_w - min_w))，其中 max_w 是最大边权重，min_w 是最小边权重。每次判断是否存在瓶颈生成树的最大权重边权重不
#     超过 b 的操作是线性时间复杂度 O(E)。所以整个算法的时间复杂度是 O(Elog(max_w - min_w))。如果边的权重范围是合理的，这个时间复杂度可以认
#     为是线性时间复杂度的。
```

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


```python
# add your idea here
    1.构建图模型：将城市视为图的顶点，道路视为图的边。每条边有一个权重，表示道路的长度。
    2.计算所有顶点对之间的最短路径：使用 Floyd-Warshall 算法计算所有顶点对之间的最短路径。这个算法的时间复杂度是 O(V^3)，其中 V 是顶点的数量。
    3.分析新道路的影响：对于每对可能的新道路 (u,v)，计算在没有这条新道路时，从 u 到 v 的最短路径长度。计算如果添加这条新道路，从u到v的最短
    路径长度（即新道路的长度l′(u,v)）。
    4.选择最优新道路：对于每对可能的新道路，计算添加这条新道路后，从 u 到 v 的距离减少量。选择使得距离减少量最大的新道路。
```

```python
# add your code here
import sys

def floyd_warshall(graph, V):
    dist = [[0 if i == j else sys.maxsize for j in range(V)] for i in range(V)]
    for i in range(V):
        for (j, weight) in graph[i]:
            dist[i][j] = weight

    for k in range(V):
        for i in range(V):
            for j in range(V):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

def find_max_reduction(graph, new_roads):
    V = len(graph)
    dist = floyd_warshall(graph, V)

    max_reduction = 0
    best_pair = (-1, -1)

    for u, v, length in new_roads:
        if dist[u][v] > length:
            reduction = dist[u][v] - length
            if reduction > max_reduction:
                max_reduction = reduction
                best_pair = (u, v)

    return best_pair, max_reduction

# 示例图
graph = [
    [(1, 3), (2, 9)],
    [(0, 3), (2, 5)],
    [(0, 9), (1, 5)],
    [(2, 2)]
]

# 新道路列表
new_roads = [(0, 1, 2), (1, 2, 4), (0, 2, 7)]

best_pair, max_reduction = find_max_reduction(graph, new_roads)
print("最佳新道路连接城市:", best_pair)
print("最大距离减少量:", max_reduction)

```

```python
# 算法正确性和复杂度分析
    算法通过计算所有顶点对之间的最短路径，并分析每条新道路对这些最短路径的影响，从而找到最优的新道路。故算法是正确的。
    主要时间开销来自于Floyd-Warshall 算法，其时间复杂度为 O(V^3)。对于每条新道路，计算其影响的时间复杂度为 O(E′)，其中E′是新道路的数量。
    因此，总的时间复杂度为 O(V^3+E′)。
```

## 问题 4

**逃离问题**

一个 $ n \times n $ 的网格是一个无向图，由 $ n $ 行和 $ n $ 列的顶点组成，如下图所示。我们用 $(i,j)$ 表示第 $ i $ 行和第 $ j $ 列的顶点。除了边界顶点，网格中的所有顶点都有四个邻居，即满足 $ i = 1, i = n, j = 1 $ 或 $ j = n $ 的点 $(i,j)$。

给定网格中的 $ m \leq n^2 $ 个起点 $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$，逃离问题是确定是否存在 $ m $ 条顶点不相交的路径（即路径之间不相交），从这些起点到边界上的任意 $ m $ 个不同点。例如，图1中的网格存在逃离。

(1) 该问题可以看作是一个最大流问题。考虑一个流网络，其中顶点和边都有容量。也就是说，进入任何给定顶点的总正流量受到容量限制。证明在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。更准确地说，你需要将一个具有顶点和边容量的网络 $ G = (V,E) $ 转换为另一个仅具有边容量的网络 $ G' = (V', E') $，使得两个网络上的最大流相同，并且你构建的新网络具有 $ V' = O(V) $ 个顶点和 $ E' = O(E) $ 条边。你可以假设网络是连通的。

(2) 描述一个解决逃离问题的高效算法，并分析其运行时间。

```python
# add your idea here
    首先构建流网络：创建一个超级源点s和一个超级汇点t。从超级源点s到每个起点连接一条容量为1的边。从每个终点到超级汇点t连接一条容量为1的边。
    网格中的每个内部顶点 (i,j) 到其四个邻居（如果存在）连接容量为 1 的边。
    然后使用最大流算法（如 Ford-Fulkerson 算法或 Edmonds-Karp 算法）计算从超级源点s到超级汇点t的最大流。如果最大流的值等于起点的数量m，则
    存在 m 条顶点不相交的路径；否则，不存在。
```

```python

    
# add your code here
import networkx as nx


def escape(n, sources, sinks):
    # 创建图
    G = nx.DiGraph()

    # 添加超级源点和超级汇点
    s = 's'
    t = 't'
    G.add_node(s)
    G.add_node(t)

    # 从超级源点到起点添加边
    for source in sources:
        G.add_edge(s, source, capacity=1)

    # 从终点到超级汇点添加边
    for sink in sinks:
        G.add_edge(sink, t, capacity=1)

    # 添加网格内部边
    for i in range(n):
        for j in range(n):
            node = (i, j)
            G.add_node(node)

            # 上边
            if i > 0:
                G.add_edge((i - 1, j), node, capacity=1)
            # 下边
            if i < n - 1:
                G.add_edge(node, (i + 1, j), capacity=1)
            # 左边
            if j > 0:
                G.add_edge((i, j - 1), node, capacity=1)
            # 右边
            if j < n - 1:
                G.add_edge(node, (i, j + 1), capacity=1)

    # 计算最大流
    max_flow = nx.maximum_flow_value(G, s, t)

    # 判断是否存在 m 条顶点不相交的路径
    return max_flow == len(sources)


# 示例
n = 5
sources = [(0, 0), (0, 1), (0, 2)]
sinks = [(n - 1, n - 1), (n - 1, n - 2), (n - 1, n - 3)]
print(escape(n, sources, sinks))  # 输出: True 或 False
```

```python
#运行时间分析
    构建图的时间复杂度为 O(n^2)，因为我们需要为每个网格顶点添加边。使用 Edmonds-Karp 算法，时间复杂度为 O(V⋅E^2)，其中 V 是顶点数，E 是边数。
    对于 n×n 的网格，顶点数 V=O(n^2)，边数 E=O(n^2)。因此，总的时间复杂度为 O((n^2)⋅(n^2)^2)=O(n^6)。
```
